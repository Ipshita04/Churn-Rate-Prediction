## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df.iloc[:,3:-1],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
X_train
```

## Encoding categorical values

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse_output=False)
```

```python
X_train['Gender']=ohe.fit_transform(X_train[['Gender']])
X_test['Gender']=ohe.transform(X_test[['Gender']])
```

```python
X_train
```

```python
geo_train_encoded = ohe.fit_transform(X_train[['Geography']])
geo_test_encoded = ohe.transform(X_test[['Geography']])
geo_train_df = pd.DataFrame(geo_train_encoded, columns=ohe.get_feature_names_out(['Geography']), index=X_train.index)
geo_test_df = pd.DataFrame(geo_test_encoded, columns=ohe.get_feature_names_out(['Geography']), index=X_test.index)
X_train = X_train.drop('Geography', axis=1)
X_test = X_test.drop('Geography', axis=1)
X_train = pd.concat([X_train, geo_train_df], axis=1)
X_test = pd.concat([X_test, geo_test_df], axis=1)
```

```python
X_train
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```
